```python
import pandas as pd
```

```python
df1= pd.read_csv("overall_hn_ratios.csv")
```

```python
df1.head()
```

```python
df2= pd.read_csv("hn_ratio_overll_2.csv")
```

```python
df2.head()
```

```python
combined_df = pd.concat([df1, df2], axis=1)  # axis=1 for columns
```

```python
combined_df.head()
```

```python
combined_df.info
```

```python
import numpy as np
```

```python
combined_df= combined_df.set_index('GeneID')
combined_df.head()
```

```python
temp = pd.DataFrame()
thre = combined_df>=2
temp["up2"] = thre.sum(axis=1)

thre = combined_df>=5
temp["up5"] = thre.sum(axis=1)
thre = combined_df>=10
temp["up10"] = thre.sum(axis=1)

thre = combined_df<=1/2
temp["dn2"] = thre.sum(axis=1)

thre = combined_df<=1/5
temp["dn5"] = thre.sum(axis=1)
thre = combined_df<=1/10
temp["dn10"] = thre.sum(axis=1)

thre = combined_df >= 0
temp["all"] = thre.sum(axis=1)

temp["unchange2"] = temp["all"] - temp["up2"] - temp["dn2"]
temp["unchange5"] = temp["all"] - temp["up5"] - temp["dn5"]
temp["unchange10"] = temp["all"] - temp["up10"] - temp["dn10"]

temp["HN2"] = temp["up2"] - temp["dn2"]
temp["HN5"] = temp["up5"] - temp["dn5"]
temp["HN10"] = temp["up10"] - temp["dn10"]

temp
```

```python
temp.to_csv("HN-score_drougt_cicer.csv", sep=",")
```
